# Prática 03 - Agentes Inteligentes

## Agente Reativo Simples: Robô Aspirador de Pó

```python
# Bibliotecas necessarias

import time
import random
import numpy as np
from random import randint
```

```python
# Definindo variaveis

agente = 'A'
agenteb = 'B'

LIMPO = 0
SUJO = 1

```

```python
def criar_ambiente():
    ambiente = [[random.choice([0, 1]) for _ in range(5)] for _ in range(5)]
    return ambiente

criar_ambiente()
```

```python
def mostrar_na_tela(ambiente, pos_agente, nome_agente):
    for r in range(5):
        linha_visual = []
        for c in range(5):
            if (r, c) == pos_agente:
                linha_visual.append(nome_agente)
            else:
                linha_visual.append(ambiente[r][c])
        print(linha_visual)
    print("\n")
```

```python
def ambiente_limpo(ambiente):
    for linha in ambiente:
        if SUJO in linha:
            return False
    return True
```

```python
def calcular_distancia_euclidiana(p1, p2):
    return np.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)
```

```python
def mover_agente(direcao, linha, coluna):

    if direcao == "cima" and linha > 0:
        linha -= 1

    elif direcao == "baixo" and linha < 4:
        linha += 1

    elif direcao == "esquerda" and coluna > 0:
        coluna -= 1

    elif direcao == "direita" and coluna < 4:
        coluna += 1

    return linha, coluna
```

```python
def rodar_agente_a(ambiente_original):

    print("--- INICIANDO AGENTE A ---")

    ambiente = [linha[:] for linha in ambiente_original]

    ns = sum(linha.count(SUJO) for linha in ambiente)

    linha = 2
    coluna = 2

    tempo = 0

    direcoes = ["cima", "baixo", "esquerda", "direita"]

    while not ambiente_limpo(ambiente):

        tempo += 1

        # limpar posição atual
        if ambiente[linha][coluna] == SUJO:
            ambiente[linha][coluna] = LIMPO

        mostrar_na_tela(ambiente, (linha, coluna), agente)

        time.sleep(0.2)

        # escolher direção aleatória
        direcao = random.choice(direcoes)

        linha, coluna = mover_agente(direcao, linha, coluna)

    te = ns / tempo if tempo > 0 else 0

    print(f"Agente A finalizou. Tempo: {tempo}, TE: {te:.4f}\n")

    return te
```

```python
def rodar_agente_b(ambiente_original):
    print("--- INICIANDO AGENTE B (REATIVO) ---")
    ambiente = [linha[:] for linha in ambiente_original]

    ns = sum(linha.count(SUJO) for linha in ambiente)
    pos_atual = [2, 2]
    tempo = 0

    while not ambiente_limpo(ambiente):
        sujeiras = [
            (r, c)
            for r in range(5)
            for c in range(5)
            if ambiente[r][c] == SUJO
        ]

        if not sujeiras:
            break

        alvo = min(sujeiras, key=lambda s: calcular_distancia_euclidiana(pos_atual, s))

        # Movimento em 4 direções
        if pos_atual[0] < alvo[0]:
            pos_atual[0] += 1
        elif pos_atual[0] > alvo[0]:
            pos_atual[0] -= 1
        elif pos_atual[1] < alvo[1]:
            pos_atual[1] += 1
        elif pos_atual[1] > alvo[1]:
            pos_atual[1] -= 1

        tempo += 1

        if ambiente[pos_atual[0]][pos_atual[1]] == SUJO:
            ambiente[pos_atual[0]][pos_atual[1]] = LIMPO

        mostrar_na_tela(ambiente, tuple(pos_atual), agenteb)
        time.sleep(0.2)

    te = ns / tempo if tempo > 0 else 0
    print(f"Agente B finalizou. Tempo: {tempo}, TE: {te:.4f}\n")
    return te
```

```python
grid_inicial = criar_ambiente()
eficiencia_a = rodar_agente_a(grid_inicial)
eficiencia_b = rodar_agente_b(grid_inicial)

print(f"Resultado Final: Agente {'A' if eficiencia_a > eficiencia_b else 'B'} é mais eficiente!")
```
